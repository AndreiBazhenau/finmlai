# Objectives
**Exploring stocks data using vectorization without looping**
- Trading Rules:
    - Stocks above 200 MA 
    - Entry when 10-period Stochastic (%K) is below 5
    - Place buy 10% limit order valid for 10 days
    - Exit on higher close OR after 10 days

```python
# Imports
import yfinance as yf
import ta 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Download Data & Indicators

```python
def indicators(stock_df):
    stock_df["SMA_200"] = ta.trend.sma_indicator(stock_df.Close, window=200)
    stock_df["stoch"] = ta.momentum.stoch(stock_df.High, stock_df.Low, stock_df.Close, window=10)
    stock_df.dropna(inplace=True)
```

```python
stock_df = yf.download("AAPL", start="2020-01-01") # download needs to be alone without showing the dataframe otherwise error of ticker not found
```

```python
# adding technical indicators to the dataframe
indicators(stock_df)
stock_df 
```

## Data Manipulation & Setting Buy Conditions

```python
# Buying conditions
stock_df["Buy"] = (stock_df.Close > stock_df.SMA_200) & (stock_df.stoch < 5) # add a new column "Buy"
stock_df[stock_df.Buy] # to see the rows where the condition is true
```

```python
# Setting the Buy Limit
stock_df["bPrice"] = np.where(stock_df.Buy, stock_df.Close * 0.97, np.nan) # if the condition is true, buy at 97% before close price instead of 100% (fully closed) , if fault set to nan
# for instance on 2020-10-28, a buy order will be issued at 107.86
stock_df.bPrice = stock_df.bPrice.ffill() # fill the nan values with the previous value
```

## Data Manipulation Setting Sell Conditions

```python
# Selling conditions
stock_df["sPrice"] = stock_df.Open.shift(-1) # shift the Open price 1 row back Why?????
stock_df
```

## Data Manipulation Shifted Low Conditions
- Limit order should check next 10 days for lower price

```python
# Goal is to have the next 10 days in one single row
for i in range(1,11): # loop through the next 10 days
    stock_df["shifted_Low_" +str(i)] = stock_df.Low.shift(-i)  # shift the Low price 1 row back 
    stock_df["shifted_Close_" +str(i)] = stock_df.Close.shift(-i) # and the Close price 
# create a list of the column names 
colnames_low = ["shifted_Low_" +str(i) for i in range(1,11)]  
colnames_close = ["shifted_Close_" +str(i) for i in range(1,11)]  
stock_df.head(7) 
# shifted_low_6 value of 114.28 is the lowest price of the (2020-10-23 row,Low column) notice that it is the same value
```

## Data Filtering & Seeing Actual Buy Conditions
- Compare If price is lower or equal to bPrice then buy

```python
raw_sig = stock_df[stock_df.Buy] # filter the Buy columns only
buy_check = raw_sig[colnames_low].le(raw_sig.bPrice, axis=0) # filter the low columns where the condition is low or equal to the bPrice using (le), axis=0 means compared to the rows 
# row of 2020-11-23 is all False so there wont be any limit order 
# the challenge is to find all the True values in the FIRST ROW (shifted_Low_1) since the condition is to buy the low if it hits the bPrice (Buy Limit). So set variable buy_check
buy_check
```
